```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import unittest


class LoginTest(unittest.TestCase):
    def setUp(self):
        options = webdriver.ChromeOptions()
        options.add_argument("--headless")  # Run in headless mode
        self.driver = webdriver.Chrome(options=options)
        self.driver.implicitly_wait(10)
        self.base_url = "https://academy.powerlearnprojectafrica.org/login"

    def test_valid_login(self):
        driver = self.driver
        driver.get(self.base_url)
        wait = WebDriverWait(driver, 10)

        # Enter valid credentials
        wait.until(EC.presence_of_element_located((By.ID, "username"))).send_keys("validuser@example.com")
        driver.find_element(By.ID, "password").send_keys("ValidPassword123")
        driver.find_element(By.XPATH, "//button[@type='submit']").click()

        # Assert successful login by checking dashboard visibility
        self.assertTrue(wait.until(EC.presence_of_element_located((By.CLASS_NAME, "dashboard"))).is_displayed())

    def test_invalid_login(self):
        driver = self.driver
        driver.get(self.base_url)
        wait = WebDriverWait(driver, 10)

        # Enter invalid credentials
        wait.until(EC.presence_of_element_located((By.ID, "username"))).send_keys("invaliduser@example.com")
        driver.find_element(By.ID, "password").send_keys("WrongPassword")
        driver.find_element(By.XPATH, "//button[@type='submit']").click()

        # Assert login failure by checking error message presence
        self.assertTrue(wait.until(EC.presence_of_element_located((By.CLASS_NAME, "error-message"))).is_displayed())

    def tearDown(self):
        self.driver.quit()


if __name__ == "__main__":
    import sys
    import io

    # Redirect stdout to capture test results
    buffer = io.StringIO()
    sys.stdout = buffer

    # Run the tests
    unittest.main(argv=['first-arg-is-ignored'], exit=False)

    # Restore stdout and print the captured output
    sys.stdout = sys.__stdout__
    print(buffer.getvalue())
```

AI-driven testing improves coverage by expanding the set of test scenarios beyond manual scripts, prioritizing high-risk paths, and adapting to UI changes automatically. It uses data from past runs, models risks, and explores edge cases that humans may overlook, leading to broader and more reliable test coverage. It also enables self-healing tests, faster test generation, and smarter prioritization, all of which boost overall test effectiveness and reduce maintenance overhead.

Key mechanisms
- Test generation and coverage expansion
  - AI analyzes app behavior and user journeys to propose new test scenarios that fill gaps in existing suites.
  - It can automatically cover rarely exercised paths (e.g., error states, boundary conditions) that are often missed in manual testing.
- Risk-based prioritization
  - By learning from past defects and test outcomes, AI prioritizes tests targeting high-risk areas, accelerating feedback on the most impactful parts of the app.
- Self-healing and stability
  - When UI changes occur, AI can adapt selectors and workflow logic, reducing brittle test failures and the need for constant script updates.
- Data-driven test design
  - AI leverages historical test data to craft varied input combinations, increasing test diversity without manual scripting effort.
- Visual and performance insights
  - Some AI-enabled tools assess visual changes and performance patterns, identifying regressions that traditional script-based tests might miss.

Practical impact
- Faster test generation and execution within CI/CD, shortening release cycles.
- Higher defect detection rates earlier in development, improving software quality.
- Lower maintenance costs due to self-healing capabilities and reduced flakiness.
- Greater test coverage across complex, data-heavy, or dynamic UI applications.
